```python
from pyspark import SparkContext
import os
os.environ["SPARK_HOME"] = "/home/hadoop/spark-2.0.1-bin-hadoop2.7"   #KeyError: 'SPARK_HOME'
sc=SparkContext(appName='Test')
```

```python
train_rdd = sc.textFile('data/train_slope.csv')
u_item_pair = train_rdd.map(lambda x:x.split(',')[0:3]).map(lambda x:(x[0],(x[1],x[2])))
```

```python
u_item_pair.take(5)
```

```python
t = u_item_pair.groupByKey()
t.take(5)
```

```python
v = u_item_pair.groupByKey().values()
v.take(5)
```

```python
mp = v.map(lambda x:list(x))
mp.take(1)
```

```python
def item_ij1(x):
    
    n = len(x)
    res = []
    for i in range(n):
        for j in range(i+1,n):
            #res.append((x[i][0],x[j][0]))
            #res.append((x[j][0],x[i][0]))
            res.append(((x[i][0],x[j][0]),(1,int(x[i][1])-int(x[j][1]))))
            res.append(((x[j][0],x[i][0]),(1,int(x[j][1])-int(x[i][1]))))
    return res

flatmp = mp.flatMap(lambda x:item_ij1(x))
flatmp.take(15)
```

```python
devij = flatmp.reduceByKey(lambda x,y:(x[0]+y[0],x[1]+y[1]))
devij.take(5)
```
